```python
class Mitochondria:

  def __init__(self, ATP, ADP):
    self.ATP = ATP
    self.ADP = ADP

  def create_ATP(self):
    print('ATP created')

  def calculate_ATP_perc(self):
    return self.ATP / self.ADP * 100
```

```python
my_mitochondria = Mitochondria(13, 14)
print(my_mitochondria.calculate_ATP_perc())
```

```python
class Cell:

  def apopthosis(self, Ca):
    if Ca >= 0.5:
      print('Start apopthosis')
    else:
      print('No apopthosis')
```

```python
class Mitochondria(Cell):

  def __init__(self, ATP, ADP):
    self.ATP = ATP
    self.ADP = ADP

  def create_ATP(self):
    print('ATP created')

  def calculate_ATP_perc(self):
    return self.ATP / self.ADP * 100
```

```python
my_mitochondria = Mitochondria(13, 14)
my_mitochondria.apopthosis(1)
```
